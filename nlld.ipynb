```python
import torch
import nltk
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# Load the pre-trained model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")


# Open the source text file for reading line by line
#source_file_path = "eng_source.txt"
#reference_file_path = "hau_reference.txt"  # Replace with the path to your reference translation file

source_file_path = "mafand-eng.txt"
reference_file_path = "mafand-hau.txt"

with open(source_file_path, "r", encoding="utf-8") as source_file, open(reference_file_path, "r", encoding="utf-8") as reference_file:
    for source_line, reference_line in zip(source_file, reference_file):
        # Process each line of source text
        source_article = source_line.strip()  # Remove leading/trailing whitespace if needed
        reference_translation = reference_line.strip()  # Remove leading/trailing whitespace if needed

        inputs = tokenizer(source_article, return_tensors="pt")

        # Replace "fra_Latn" with the appropriate language code
        target_language_code = "hau_Latn"
        forced_bos_token_id = tokenizer.lang_code_to_id[target_language_code]

        # Generate the translated tokens
        translated_tokens = model.generate(
            **inputs, forced_bos_token_id=forced_bos_token_id, max_length=100
        )

        # Decode the translated tokens
        translated_sentence = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
        print("Translated:", translated_sentence)

        # Calculate BLEU score
        reference_tokenized = nltk.word_tokenize(reference_translation)
        translated_tokenized = nltk.word_tokenize(translated_sentence)
        bleu_score = nltk.translate.bleu_score.sentence_bleu([reference_tokenized], translated_tokenized)
        print("BLEU Score:", bleu_score)
```

```python
import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import nltk

# Load the pre-trained model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")

# Open the source text file and reference translation file for reading
#source_file_path = "eng_source.txt"
#reference_file_path = "hau_reference.txt"

source_file_path = "mafand-eng.txt"
reference_file_path = "mafand-hau.txt"

source_lines = []
reference_lines = []

with open(source_file_path, "r", encoding="utf-8") as source_file, open(reference_file_path, "r", encoding="utf-8") as reference_file:
    source_lines = source_file.readlines()
    reference_lines = reference_file.readlines()

# Initialize a list to store BLEU scores for each instance
bleu_scores = []

for source_line, reference_line in zip(source_lines, reference_lines):
    # Process each line of source text
    source_article = source_line.strip()  # Remove leading/trailing whitespace if needed
    reference_translation = reference_line.strip()  # Remove leading/trailing whitespace if needed

    inputs = tokenizer(source_article, return_tensors="pt")

    # Replace "fra_Latn" with the appropriate language code
    target_language_code = "hau_Latn"
    forced_bos_token_id = tokenizer.lang_code_to_id[target_language_code]

    # Generate the translated tokens
    translated_tokens = model.generate(
        **inputs, forced_bos_token_id=forced_bos_token_id, max_length=100
    )

    # Decode the translated tokens
    translated_sentence = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]

    # Calculate BLEU score
    reference_tokenized = nltk.word_tokenize(reference_translation)
    translated_tokenized = nltk.word_tokenize(translated_sentence)
    bleu_score = nltk.translate.bleu_score.sentence_bleu([reference_tokenized], translated_tokenized)
    
    # Append the BLEU score to the list
    bleu_scores.append(bleu_score)

# Calculate the average BLEU score
average_bleu = sum(bleu_scores) / len(bleu_scores)
print("Average BLEU Score:", average_bleu)

#Average BLEU Score: 0.16800131210642766
```

```python
import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import nltk
import pandas as pd  # Import pandas for reading TSV files
from nltk.translate.bleu_score import SmoothingFunction

# Create a SmoothingFunction object
smoothing = SmoothingFunction()

# Load the pre-trained model and tokenizer
#tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
#model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-1.3B")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-1.3B")

#facebook/nllb-200-3.3B

# Specify the path to your TSV file with source text and translations
tsv_file_path = "mafand-test-hausa.tsv"  # Replace with the actual path to your TSV file

# Load the TSV file into a pandas DataFrame
df = pd.read_csv(tsv_file_path, sep="\t", header=None, names=["source_text", "reference_translation"])

# Initialize a list to store BLEU scores for each instance
bleu_scores = []

for _, row in df.iterrows():
    source_article = row["source_text"].strip()  # Remove leading/trailing whitespace if needed
    reference_translation = row["reference_translation"].strip()  # Remove leading/trailing whitespace if needed

    inputs = tokenizer(source_article, return_tensors="pt")

    # Replace "hau_Latn" with the appropriate language code
    target_language_code = "hau_Latn"
    forced_bos_token_id = tokenizer.lang_code_to_id[target_language_code]

    # Generate the translated tokens
    translated_tokens = model.generate(
        **inputs, forced_bos_token_id=forced_bos_token_id, max_length=100
    )

    # Decode the translated tokens
    translated_sentence = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]

    # Calculate BLEU score
    reference_tokenized = nltk.word_tokenize(reference_translation)
    translated_tokenized = nltk.word_tokenize(translated_sentence)
    #bleu_score = nltk.translate.bleu_score.sentence_bleu([reference_tokenized], translated_tokenized)
    bleu_score = nltk.translate.bleu_score.sentence_bleu([reference_tokenized], translated_tokenized, smoothing_function=smoothing.method1)

    # Append the BLEU score to the list
    bleu_scores.append(bleu_score)

# Calculate the average BLEU score
average_bleu = sum(bleu_scores) / len(bleu_scores)
print("Average BLEU Score:", average_bleu)
```
