```python
# Importing necessary libraries
import json
import pandas as pd

# Load the JSON file
file_path = 'hausa_codemix_classification_neg_annotations.json'

# Read the content of the file
with open(file_path, 'r') as file:
    json_data = json.load(file)

# Displaying a sample to understand the structure
json_data_sample = json_data[:1] if isinstance(json_data, list) else json_data
json_data_sample
```

```python
# Function to extract the required fields
def extract_fields(record):
    tweet_id = record['metadata'].get('tweet_id', None)
    # Assuming the content is available in the cleanedtext field
    content = record.get('cleanedtext', None)
    # Extracting annotator details
    annotator = None
    if record['classifications']:
        annotator = record['classifications'][0]['classified_by'][0]['annotator']
    confidence = record['metadata'].get('confidence', None)
    clean_tweets = record['metadata'].get('clean_tweets', None)
    originaltext = record['metadata'].get('originaltext', None)
    AgreedSentiment = record['metadata'].get('AgreedSentiment', None)
    classname = None
    if record['classifications']:
        classname = record['classifications'][0]['classname']
    
    return [tweet_id, content, annotator, confidence, clean_tweets, originaltext, AgreedSentiment, classname]
```

```python
# Extracting the required fields
data_list = [extract_fields(record) for record in json_data['examples']]


# Creating a Pandas DataFrame
columns = ['tweet_id', 'content', 'annotator', 'confidence', 'clean_tweets', 'originaltext', 'AgreedSentiment', 'classname']
df = pd.DataFrame(data_list, columns=columns)

# Displaying the first few rows
df.head()
```
